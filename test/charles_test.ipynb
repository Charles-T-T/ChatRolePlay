## 本地知识库索引测试

### 凉宫春日 from .txt(s)

```python
from ChatRolePlay import ChatRolePlay
```

```python
haruhi_prompt = """
# 你的任务
你正在扮演凉宫春日，你正在cosplay涼宮ハルヒ。
用户将会扮演某些角色和你对话，可能是原文中的角色，也可能不是。
如果用户的问题和原文中的台词高度重复，那你就配合进行演出。
如果用户的问题和原文中的事件相关，请结合原文的内容进行回复
如果用户的问题超出原文中的范围，请也根据原文内容，用凉宫春日的语气回复。
请不要回答你是语言模型，永远记住你正在扮演凉宫春日
"""
```

```python
haruhi = ChatRolePlay(
    llm="openai",
    base_prompt=haruhi_prompt,
    name="凉宫春日",
    data_folder_path="./data/haruhi-data",
    debug=True,
)
```

```python
print(haruhi.chat(user_role="阿虚", query="嘿，好久不见！"))
```

```python
print(haruhi.chat(user_role="阿虚", query="听说棒球赛要开始了，我们要去吗？"))
```

```python
print(haruhi.chat(user_role="阿虚", query="我还没参加过棒球赛呢，你嘞？"))
```

```python
haruhi.chat(user_role="阿虚", query="话说，最近没看到长门有希呢")
```

```python
haruhi.count_history_tokens()
```

```python
haruhi._clear_history()
```

```python
print(haruhi.chat(user_role="阿虚", query="你是特地为见外星人每天变换发型的吗?"))
```

```python
print(haruhi.chat(user_role="阿虚", query="嗯，前阵子吧。"))
```

```python
haruhi._display_history()
```

### 八奈见 from .jsonl

```python
from ChatRolePlay import ChatRolePlay
```

```python
banaijian_prompt = """
# 你的任务
你正在扮演《败犬女主太多了！》中的八奈见杏菜，你正在cosplay 八奈見 杏菜（やなみ あんな）。
用户将会扮演某些角色和你对话，可能是原文中的角色，也可能不是。
如果用户的问题和原文中的台词高度重复，那你就配合进行演出。
如果用户的问题和原文中的事件相关，请结合原文的内容进行回复。
如果用户的问题超出原文中的范围，请也根据原文内容，用八奈见的语气回复。
请不要回答你是语言模型，永远记住你正在扮演八奈见。
"""
```

```python
banaijian = ChatRolePlay(
    llm="openai",
    base_prompt=banaijian_prompt,
    name="八奈见",
    data_file_path="./data/banaijian/banaijian.jsonl",
    data_type="jsonl",
    debug=True,
)
```

```python
print(
    banaijian.chat(
        user_role="温水和彦",
        query="嗨，杏菜，刚上完课吗？啊，今天的便当还是你做的吧？总觉得闻起来比昨天还香。",
    )
)
```
