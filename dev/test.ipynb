# JSON 本地抽取测试

```python
banaijian_data_path = "./data/banaijian.jsonl"
chat_banai = ChatRolePlay(
    role_from_jsonl=banaijian_data_path,
    llm="openai"
) 
```

```python
chat_banai.chat(role="小鞠", text="你好呀，八奈")
```

```python
chat_banai.system_prompt
```

```python
query = chat_banai.get_query_string("小鞠", "你好呀，八奈")
query_vec = chat_banai.embedding(query)
chat_banai.db.search(query_vec, n_results=chat_banai.k_search)
```
