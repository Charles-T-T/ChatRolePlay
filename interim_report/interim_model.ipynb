```python
def get_bg(model, role, query):
    """根据用户的query抽取的相关故事背景"""
    query = model.get_query_string(role=role, text=query)
    query_vec = model.embedding(query)
    bg = model.db.search(query_vec, model.k_search)
    return bg


bg = get_bg(banaijian, "温水", "八奈见同学，你好啊")
for story in bg:
    if "扣子" in story:
        print(story)
```
