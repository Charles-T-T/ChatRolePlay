```python
import re


def split_by_chapter(input_file):
    # 读取西游记文本
    with open(input_file, "r", encoding="utf-8") as f:
        text = f.read()

    # 定义正则表达式以识别回目章节，例如："第一回"、"第二回"等
    chapter_pattern = re.compile(
        r"第[一二三四五六七八九十百千]+回\s+.*?[\n\r]", re.DOTALL
    )

    # 查找所有回目的开始位置
    chapters = chapter_pattern.findall(text)

    # 如果没有找到回目，提示并退出
    if not chapters:
        print("没有找到任何回目！")
        return

    # 根据回目切分文本
    chapter_start = 0
    for i, chapter in enumerate(chapters):
        chapter_end = text.find(chapters[i + 1]) if i + 1 < len(chapters) else len(text)

        chapter_text = text[chapter_start:chapter_end].strip()
        chapter_start = chapter_end

        # 保存每一回为单独的文件
        chapter_filename = f"chap_{i+1}.txt"
        with open(chapter_filename, "w", encoding="utf-8") as chapter_file:
            chapter_file.write(chapter_text)

        print(f"已保存第{i+1}回内容到 {chapter_filename}")


# 执行函数，传入西游记的文本文件路径
split_by_chapter("西游记.txt")
```
